```python
store1={'хлеб':105,'молоко':65,'syr':120,'yaico':80,'chai':150}
store2={'хлеб':789,'yogurt':70,'syr':90,'smetana':87,'chai':890}
store1.update(store2)
print(store1)
for key in store2:
    if key in store1:
        store1[key]=max(store1[key],store2[key])
    else:
        store1[key]=store2[key]
print(store1)
```
