```python
bank = {
    "Checking": [
        {"id": "C001", "holder": "Alice Kim",   "balance": 1200.00, "currency": "USD"},
        {"id": "C002", "holder": "B. Lee",      "balance":  -50.00, "currency": "USD"},
        {"id": "C003", "holder": "Carlos Diaz", "balance":  500.00, "currency": "KGS"},
        {"id": "C004", "holder": "D. Singh",    "balance": 3000.00, "currency": "USD"},
        {"id": "C005", "holder": "E. Toktogul", "balance":15000.00, "currency": "KGS"}
    ]
 }
```

```python


def  avg_balance(bank: dict) -> dict[str, float]:
    sum_usd=0
    sum_kg=0
    count=0
    count1=0
    for Checking in bank.values():
        for s in Checking:
            if s["currency"]=="USD":
                sum_usd+=s["balance"]
                count+=1
            elif s["currency"]=="KGS":
                sum_kg+=s["balance"]
                count1+=1
        usd_avr=sum_usd/count
        kg_avr=sum_kg/count1
    return { "USD":usd_avr, "KGZ":kg_avr}
print(avg_balance(bank))


 
                
```

```python
def count_accounts(bank: dict) -> int:
    count=0
    for checking in bank.values():
        for s in checking:
            if s in checking:
                count+=1
        return count
print(count_accounts(bank))
```

```python
products = {
    "001": {"name": "pen", "price": 10},
    "002": {"name": "notebook", "price": 50},
    "003": {"name": "eraser", "price": 5}
}
```

```python
def lookup(ima):
    item=products.get(ima, None)
    try:
       name=item['name'] 
       price=item['price']
       return f'{ima}:{name}->{price}'
    except:
        return f'{ima} не найдено'
    
print(lookup('001'))
```
