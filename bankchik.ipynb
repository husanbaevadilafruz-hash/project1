```python
def top_k_accounts(bank, k: int) -> list[dict]:
    index_id = find_index(bank[0],"account_id")
    index_holder = find_index(bank[0], "holder")
    index_balance = find_index(bank[0], "balance")
    index_currency = find_index(bank[0],"currency")
    dicshinary={}
    for schet in bank[1:]:
        valuta=schet[index_currency]
        record = {
            "account_id": schet[index_id],
            "holder": schet[index_holder],
            "balance": float(schet[index_balance]),
            "currency": valuta
        }
        if valuta not in dicshinary:
           dicshinary[valuta]=[]
        dicshinary[valuta].append(record)
    result=[]
    for valuta, accounts in dicshinary.items():
        sorted_accounts=sorted(accounts, key=lambda x: x['balance'], reverse=True)
        result.extend(sorted_accounts[:k])
    return result
print(top_k_accounts(bank, 1))
```

```python
import csv
 
def read_csv(filename):
    with open(filename, 'r', encoding='utf-8', newline='') as f:
        reader=csv.DictReader(f)
    return list(reader)

```

```python
def count_books(filename):
    books=read_csv(filename)
    return len(books)
```

```python
def available(filename):
    count=0
    books=read_csv(filename)
    for book in books:
        if book('available')=='True':
            count+=1
    return count
```

```python
def checkout_book(filename, title):
    find=False
    books=read_csv(filename)
    for book in books:
        if book('title')==title:
            book('available')='True'
            find=True
    if find:
        with open(filename, 'w', encoding='utf-8', newline='') as f:
            writer=csv.DictWriter(f, fieldname=['available', 'title', 'section'])
            writer.writeheader()
            writer.writerows(books)
    return find
```

```python
from flask import Flask, request
from collections import defaultdict
menu={}
zakazy_klienta=defaultdict(list)
istoria=[]
app=Flask(__name__) 
@app.route('/sozdat_menu')
def sozdat_menu():
    try:
        food=request.args.get('food')
        price=request.args.get('price')
        if not price or not food:
            return 'vvedite bludo i cenu'
        price=int(price)
        menu[food]=price 
        istoria.append(f'{food} s cenoi {price} dobavlen v menu')
    except Exception as e :
        return f'oshibka: {e}'
@app.route('/zakazat')
def zakazat():
    try:
        food=request.args.get('food')
        price=request.args.get('price')
        ima=request.args.get('ima')
        price=int(price)
        if food in menu and menu[food]==price:
            zakazy_klienta[ima].append((food, price))
            istoria.append(f'{ima}  zakazal {food}, po cene {price}')
        else:
            istoria.append(f'{ima} pytalsa zakazat {food}, no ego net')
    except Exception as e:
        return f"oshibka: {e}"

@app.route('/otmena_zakaza')
def otmena_zakaza():
    try:
        food=request.args.get('food')
        price=int(request.args.get('price'))
        ima=request.args.get('ima')
        if (food, price) in zakazy_klienta[ima]:
            zakazy_klienta[ima].remove((food, price))
            istoria.append(f'{ima} otmenil zakaz bluda {food}')
        else:
            istoria.append(f'{ima} pytalsa otmenit zakaz, kotorogo ne delal, aahhaahh')
    except Exception as e:
        return f'oshibka: {e}'   
@app.route('/istoria')
def vernut_istoriu():
    return '<br>'.join(istoria) 
if __name__=='__main__':
    app.run(debug=True)




```
