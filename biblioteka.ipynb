```python
def dedupe_section_titles(arr, section: str):
    section_index = find_index(arr[0], 'section')
    title_index=find_index(arr[0],'title')
    seen_titles=set()
    removed_titles=[]
    for book in arr[1:]:
        if book[section_index]==section:
            title=book[title_index]           #poschitat dublikaty
            if title not in seen_titles:
                seen_titles.add(title)
            if title in seen_titles:
                removed_titles.append(title)
    return len(removed_titles)
print(dedupe_section_titles(biblioteka, 'Science'))
```

```python
def otchet(arr):
    summary = {}  # создаем пустой словарь для хранения статистики
    section_index = find_index(arr[0], 'section')     # индекс колонки "section"
    available_index = find_index(arr[0], 'available') # индекс колонки "available"

    # Считаем total, available, checkout для каждого раздела
    for book in arr[1:]:  # пропускаем заголовок
        section = book[section_index].strip()         # убираем пробелы
        available = book[available_index].strip().lower()  # нормализуем "True"/"False"

        if section not in summary:  # если раздел еще не встречался
            summary[section] = {'total': 0, 'available': 0, 'checkout': 0}

        summary[section]['total'] += 1
        if available == 'true':
            summary[section]['available'] += 1
        else:
            summary[section]['checkout'] += 1

    # После подсчета считаем доли
    for section in summary:
        total = summary[section]['total']
        summary[section]['available_ratio'] = summary[section]['available'] / total
        summary[section]['checkout_ratio'] = summary[section]['checkout'] / total

    return summary

# Пример вызова
result = otchet(biblioteka)
print(result)
```

```python
def availability_ratio(arr) -> dict[str, float]:
    ratio={}
    section_index = find_index(arr[0], 'section')     
    available_index = find_index(arr[0], 'available')
    for book in arr[1:]:
        section=book[section_index]
        available=book[available_index]
        if section not in ratio:
            ratio[section]={"total":0,"available":0}
            ratio[section]['total']+=1
        if available=='True':
            ratio[section]['available']+=1


    result={}
    for section in ratio:
         result[section]=ratio[section]['available']/ratio[section]['total']
            

    return result
print(availability_ratio(biblioteka))

```

```python
def top_k_available_by_section(arr, section: str, k: int) ->list[str]:
    liste=[]
    section_index = find_index(arr[0], 'section')     
    available_index = find_index(arr[0], 'available')
    title_index=find_index(arr[0],'title')
    for book in arr[1:]:
        if (
            book[section_index]==section
            and book[available_index]=='True'
        ):
            liste.append(book[title_index])
    x=sorted(liste)[:k]
    return x
print(top_k_available_by_section(biblioteka, " History",2 ))
```

```python
def normalize_titles(arr) -> int:
    title_index=find_index(arr[0],'title')
    count=0
    for book in arr[1:]:
        old_book_title=book[title_index]
        new_book_title=''.join(old_book_title.strip().split()).title() 
        if new_book_title!=book[title_index]:
            book[title_index]=new_book_title
            count+=1
    return count
print(normalize_titles(biblioteka))
```

```python
def compare_sections(arr, section1: str, section2: str) -> list[str]:
    title_index=find_index(arr[0],'title')
    section_index = find_index(arr[0], 'section') 
    sectionn1=set()
    sectionn2=set()
    for book in arr[1:]:
        if book[section_index]==section1:
            sectionn1.add(book[title_index])
        elif book[section_index]==section2:
            sectionn2.add(book[title_index])  

    
    x=sectionn1 & sectionn2
    return sorted(x)
    return False
print(compare_sections(biblioteka, 'History', 'Science'))

```

```python
def rename_section(arr, old_section: str, new_section: str) -> int:
    section_index = find_index(arr[0], 'section') 
    count=0
    for book in arr[1:]:
        if (book[section_index]==old_section 
            and old_section!=new_section):
            book[section_index]=new_section
            count+=1
    return count
print(rename_section(biblioteka,' History',' Science'))
            
```

```python
def toggle_availability(arr, title: str) -> bool:

    available_index = find_index(arr[0], 'available')
    index_title=find_index(arr[0], 'title')
    for book in arr[1:]:
        if book[index_title]==title:
            if book[available_index]=='True':
                book[available_index]='False'
            else:
                book[available_index]='True'
            return True
    return False
print(toggle_availability(biblioteka,'Quantum Mechanics'))
print(biblioteka) 
```

```python
def duplicate_book_to_section(arr, title: str, to_section: str) ->bool:
    section_index = find_index(arr[0], 'section')
    index_title=find_index(arr[0], 'title')
    for book in arr[1:]:
        if book[index_title]==title:
            new_book=book.copy()
            new_book[section_index]=to_section
            arr.append(new_book)
            return True
    return False 
print(duplicate_book_to_section(biblioteka, 'Quantum Mechanics', 'History'))
```

```python
def sort_section_by_title(arr, section: str):
    section_index = find_index(arr[0], 'section')
    index_title=find_index(arr[0], 'title')
    listik=[]
    for book in arr[1:]:
        if book[section_index]==section:
            listik.append(book[index_title])
    return sorted(listik)
print(sort_section_by_title(biblioteka, 'History'))
```
