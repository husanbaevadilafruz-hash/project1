```python
class Circle:
    def __init__(self, radius):
        self.radius=radius
    def area(self):
        return f' ploshad kruga s radiusom {self.radius} ={3.14*self.radius**2}'
c1=Circle(4)
print(c1.area())
```

```python
# –°–æ–∑–¥–∞–π –∫–ª–∞—Å—Å Rectangle, –∫–æ—Ç–æ—Ä—ã–π:

# –ü—Ä–∏–Ω–∏–º–∞–µ—Ç —à–∏—Ä–∏–Ω—É w –∏ –≤—ã—Å–æ—Ç—É h —á–µ—Ä–µ–∑ __init__.

# –°—á–∏—Ç–∞–µ—Ç –ø–µ—Ä–∏–º–µ—Ç—Ä (2*(w+h)) –≤ –º–µ—Ç–æ–¥–µ perimeter().

# –°—á–∏—Ç–∞–µ—Ç –ø–ª–æ—â–∞–¥—å (w*h) –≤ –º–µ—Ç–æ–¥–µ area().

# –í—ã–≤–æ–¥–∏—Ç —á–µ—Ä–µ–∑ __str__: "–ü—Ä—è–º–æ—É–≥–æ–ª—å–Ω–∏–∫: —à–∏—Ä–∏–Ω–∞ = ..., –≤—ã—Å–æ—Ç–∞ = ..., –ø–µ—Ä–∏–º–µ—Ç—Ä = ..., –ø–ª–æ—â–∞–¥—å = ..."

# –ü–æ–ø—Ä–æ–±—É–π —Å–æ–∑–¥–∞—Ç—å –æ–±—ä–µ–∫—Ç —Å —à–∏—Ä–∏–Ω–æ–π 4 –∏ –≤—ã—Å–æ—Ç–æ–π 10 –∏ –≤—ã–≤–µ—Å—Ç–∏ –µ–≥–æ.
class Restangle:
    def __init__(self, dlina, shirina):
        self.dlina=dlina
        self.shirina=shirina
    def perimeter(self):
        return (self.dlina+self.shirina)*2
    def ploshad(self):
        return self.dlina*self.shirina
    def __str__(self):
        return f'perimetr= {self.perimeter()}, ploshad= {self.ploshad()}'
pr1=Restangle(4,10)
print(pr1)
```

```python
# –°–æ–∑–¥–∞–π —Ç—Ä–∏ –∫–ª–∞—Å—Å–∞: Processor, RAM, HardDrive.

# Processor –∏–º–µ–µ—Ç –∞—Ç—Ä–∏–±—É—Ç model –∏ –º–µ—Ç–æ–¥ speed() ‚Üí –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç —á–∞—Å—Ç–æ—Ç—É –≤ –ì–ì—Ü.

# RAM –∏–º–µ–µ—Ç –∞—Ç—Ä–∏–±—É—Ç size –∏ –º–µ—Ç–æ–¥ info() ‚Üí –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç —Ä–∞–∑–º–µ—Ä –ø–∞–º—è—Ç–∏.

# HardDrive –∏–º–µ–µ—Ç –∞—Ç—Ä–∏–±—É—Ç capacity –∏ –º–µ—Ç–æ–¥ info() ‚Üí –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç –æ–±—ä—ë–º –¥–∏—Å–∫–∞.

# –°–æ–∑–¥–∞–π –∫–ª–∞—Å—Å Computer, –∫–æ—Ç–æ—Ä—ã–π –≤–Ω—É—Ç—Ä–∏ —Å–µ–±—è —Å–æ–¥–µ—Ä–∂–∏—Ç processor, ram, hard_drive.

# –ú–µ—Ç–æ–¥—ã: set_processor(model, ghz), set_ram(size), set_hard_drive(capacity) ‚Üí —Å–æ–∑–¥–∞—é—Ç –æ–±—ä–µ–∫—Ç—ã —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∏—Ö –∫–ª–∞—Å—Å–æ–≤.

# –ü–æ—Å–ª–µ —ç—Ç–æ–≥–æ –º–æ–∂–Ω–æ —Å–æ–∑–¥–∞—Ç—å –æ–±—ä–µ–∫—Ç Computer, –¥–æ–±–∞–≤–∏—Ç—å –∫–æ–º–ø–æ–Ω–µ–Ω—Ç—ã –∏ –≤—ã–∑–≤–∞—Ç—å –º–µ—Ç–æ–¥—ã –∫–∞–∂–¥–æ–≥–æ –∫–æ–º–ø–æ–Ω–µ–Ω—Ç–∞ —á–µ—Ä–µ–∑ –æ–±—ä–µ–∫—Ç Computer.

class Processor:
    def __init__(self, model, chastota):
        self.model=model
        self.chastota=chastota
    def speed(self):
        return self.chastota
    def __str__(self):
        return f'model {self.model} imeet chastotu {self.chastota}'

class RAM:
    def __init__(self, size):
        self.size=size
    def info(self):
        return self.size
    def __str__(self):
        return f'{self.size}'
class HardDrive:
    def __init__(self, capacity):
        self.capacity=capacity
    def info(self):
        return self
    def __str__(self):
        return f'{self.capacity}'
class Computer:
    def __init__(self):
        self.processor=None
        self.harddrive=None
        self.ram=None
    def set_processor(self, model,chastota):
        self.processor=Processor(model,chastota)
    def set_harddrive(self, capacity):
        self.harddrive=HardDrive(capacity)
    def set_ram(self, size):
        self.ram=RAM(size)
```

```python
# === –ü–∞—Å—Å–∞–∂–∏—Ä ===
class Passenger:
    def __init__(self, name, seat=None):
        self.name = name
        self.seat = seat


# === –ë–æ—Ä—Ç–ø—Ä–æ–≤–æ–¥–Ω–∏–∫ ===
class FlightAttendant:
    def __init__(self, name):
        self.name = name

    def safety_demo(self):
        print(f"{self.name}: –ø—Ä–æ–≤–æ–¥–∏—Ç –¥–µ–º–æ–Ω—Å—Ç—Ä–∞—Ü–∏—é –±–µ–∑–æ–ø–∞—Å–Ω–æ—Å—Ç–∏.")

    def serve(self, passenger):
        print(f"{self.name}: –æ–±—Å–ª—É–∂–∏–≤–∞–µ—Ç {passenger.name}.")


# === –°–∞–º–æ–ª—ë—Ç—ã ===
class Airplane:
    def __init__(self, model, capacity):
        self.model = model
        self.capacity = capacity                   #vmestitelnost


class PrivateJet(Airplane):                             #chastnyi samolet
    def __init__(self, model, capacity, owner):
        super().__init__(model, capacity)
        self.owner = owner
        self.attendants = []                             #—Å–æ—Ç—Ä—É–¥–Ω–∏–∫–∏
        self.passengers = []

    def assign_attendants(self, attendants):              #–Ω–∞–∑–Ω–∞—á–∞—Ç—å
        self.attendants = attendants

    def order(self, client, route, passengers):
        self.passengers = passengers
        print(f"{self.model}: –∑–∞–∫–∞–∑–∞–Ω –∫–ª–∏–µ–Ω—Ç–æ–º {client} –¥–ª—è –º–∞—Ä—à—Ä—É—Ç–∞ {route[0]} ‚Üí {route[1]}")

    def fly(self):
        print(f"{self.model}: –≤—ã–ø–æ–ª–Ω—è–µ—Ç —á–∞—Å—Ç–Ω—ã–π —Ä–µ–π—Å {self.owner}.")
        for a in self.attendants:
            a.safety_demo()
        for p in self.passengers:
            print(f"{p.name} –Ω–∞—Å–ª–∞–∂–¥–∞–µ—Ç—Å—è –ø–æ–ª—ë—Ç–æ–º.")


class CommercialAirplane(Airplane):
    def __init__(self, model, capacity, airline):
        super().__init__(model, capacity)
        self.airline = airline
        self.attendants = []
        self.passengers = []

    def assign_attendants(self, attendants):
        self.attendants = attendants

    def board_passengers(self, passengers):
        self.passengers = passengers
        for p in passengers:
            print(f"{p.name} —Å–∞–¥–∏—Ç—Å—è –Ω–∞ –º–µ—Å—Ç–æ {p.seat}.")

    def fly(self):
        print(f"{self.model} –∞–≤–∏–∞–∫–æ–º–ø–∞–Ω–∏–∏ {self.airline} –≤—ã–ø–æ–ª–Ω—è–µ—Ç —Ä–µ–π—Å.")
        for a in self.attendants:
            a.safety_demo()
        for p in self.passengers:
            print(f"{p.name} –ª–µ—Ç–∏—Ç –≤ –∫—Ä–µ—Å–ª–µ {p.seat}.")


# === –ü—Ä–∏–º–µ—Ä ===
if __name__ == "__main__":
    # –ß–∞—Å—Ç–Ω—ã–π —Å–∞–º–æ–ª—ë—Ç
    jet = PrivateJet("Gulfstream G650", 10, "SkyLux")
    jet.assign_attendants([FlightAttendant("–ê–ª–∏–Ω–∞")])
    jet.order("VipTravel", ("FRU", "DXB"), [Passenger("–î–∞–Ω–∏—ç–ª—å"), Passenger("–ê–π–≥–µ—Ä–∏–º")])
    jet.fly()

    print("-" * 40)

    # –ö–æ–º–º–µ—Ä—á–µ—Å–∫–∏–π —Å–∞–º–æ–ª—ë—Ç
    plane = CommercialAirplane("Boeing 737", 180, "Air Kyrgyz")
    plane.assign_attendants([FlightAttendant("–ú–∞—Ä–∏–∞–º"), FlightAttendant("–ê–∑–∞—Ç")])
    plane.board_passengers([Passenger("–ù—É—Ä–¥–∞—É–ª–µ—Ç", "12A"), Passenger("–ú–∏—Ä–ª–∞–Ω", "12B")])
    plane.fly()
```
