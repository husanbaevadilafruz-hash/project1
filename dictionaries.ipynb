```python
def  checkout_section(library:dict, section:str)->int:
    vydannye=0
    books=library.get(section,[])
    for book in books:
        if book["available"]:                                     #выдать количество выданных книг в раз
            book["available"]=False
            vydannye+=1
    
    return vydannye
print(checkout_section(library, "Science"))
```

```python
def count_checked_out(library: dict)->int:
    count=0
    for books in library.values():
        for book in books:
            if book["available"]==False:               #посчитать количество недоступных книг
                count+=1

    return count

print(count_checked_out(library))
```

```python
def longest_title(library: dict) -> str | None:
    longest_book=None
    for books in library.values():
        for book in books:
            if longest_book is None or len(sorted(book["title"]))>len(sorted(longest_book["title"])):
                longest_book=book
    return longest_book["title"] if longest_book else None     #самая длинная книга

print(longest_title(library))
            
```

```python
def max_book_bucva(library:dict)->str|None:
     longest=None
     for books in library.values():
          for book in books:
               if(
                   longest is None                                    #самая длинная книга (алфавит)
                     or len(book["title"])>len(longest["title"])
                     or (len(book["title"])==len(longest["title"])
                     and book["title"].casefold()<longest["title"].casefold())
               ):
                   longest=book
                   return longest["title"] if longest else None 
print(max_book_bucva(library)) 
                    
```
