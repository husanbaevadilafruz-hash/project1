```python
#transpose matrix

def transpose(matrix:list):
    colums=len(matrix[0])
    rows=len(matrix)
    if colums==0:
        return []
    transposed=[[0 for _ in range(rows)]  for _ in range(colums)] 
    for i in range(rows):
        for j in range(colums):
            transposed[j][i]=matrix[i][j]
    return transposed
print(transpose([[3,4,5,2],[5,6,7,3]]))   
 
```

```python
def trans(matrix):
    m_rows = len(matrix)
    m_cols = len (matrix[0])
    if m_cols == 0:
        return []
    transposed = [[0 for _ in range(m_rows)] for _ in range (m_cols)]
    for i in m_rows:
        for j in m_cols:
            transposed[j][i] = matrix[i][j]
    return transposed
```

```python
def trans(matrix):
    rows = len(matrix)
    columns = len(matrix[0])
    if columns == 0:
        return []
    transposed = [[0 for _ in range(rows)] for _ in range(columns)]
    for i in range(rows):
        for j in range(columns):
            transposed[j][i] = matrix[i][j]
    return transposed
```
