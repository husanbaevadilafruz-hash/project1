```python
def issum(arr, target):
    result=[]
    for i in range(len(arr)):
        for j in range(i+1, len(arr)):
            if arr[i]+arr[j]==target:
                result.append([arr[i], arr[j]])
        return result
print(issum([2,34,5,6,3,1,1,4,5,36,2,7,1], 6))

    
```

```python
items = {"cola":55, "shoro": 65, "mars":40, "sprite":60, "nan":25, "burger":150, "ruchka":90}
def menshe_target(x, target):
    result=[]
    for k, v in x.items():
        if v<target:
            result.append(k)
    return result

print(menshe_target(items, 60))

def summa(x, cart):
    total=0.0
    for k, v in x.items():
        if k in cart:
            total+=v
    return total
print(summa(items, ['cola', 'mars']))

def obnova(x):
    new={}
    for k, v in x.items():
        new[k]=v-v/10
    return new
print(obnova(items))
```
