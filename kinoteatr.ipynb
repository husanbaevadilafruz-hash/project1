```python
def apply_discount(arr, movie: str|None, format: str|None, time:str|None, percent: float) -> int:
    count=0
    movie_index=find_index(arr[0],'movie')
    time_index=find_index(arr[0], 'time') 
    available_index=find_index(arr[0],'booked')
    klient_index=find_index(arr[0],'customer')
    id_index=find_index(arr[0],'id') 
    price_index=find_index(arr[0],'price')
    seat_index = find_index(arr[0], 'seat')
    format_index=find_index(arr[0], 'format')
    for film in arr[1:]:
        if (film[movie_index]==movie or movie is None) and (film[format_index]==format or format is None) and (film[time_index]==time or time is None):
            film[price_index]=float(film[price_index])
            film[price_index]-=film[price_index]*percent/100
            count+=1
    return count 
```

```python
def checkout(arr, ticket_ids: list[str], customer: str) -> float:
    
    movie_index=find_index(arr[0],'movie')
    time_index=find_index(arr[0], 'time')  
    available_index=find_index(arr[0],'booked')
    klient_index=find_index(arr[0],'customer')
    id_index=find_index(arr[0],'id') 
    price_index=find_index(arr[0],'price')
    seat_index = find_index(arr[0], 'seat')
    format_index=find_index(arr[0], 'format')
    for ima in ticket_ids:
        found=False
        for film in arr[1:]:
            if film[id_index]==ima:
                found=True
                if film[available_index]=='True':
                    return 0.0
        if not found:
            return 0.0
            
    total=0.0
    for ima in ticket_ids:
        for film in arr[1:]:
            if film[id_index]==ima:
                film[available_index]='True'
                film[klient_index]=customer
                total+=float(film[price_index])
    return total
```

```python
def swap_bookings(arr, ticket_id_a: str, ticket_id_b: str) -> bool:
    id_index = find_index(arr[0], 'id')
    time_index = find_index(arr[0], 'time')
    booked_index = find_index(arr[0], 'booked')
    customer_index = find_index(arr[0], 'customer')
    seat_index = find_index(arr[0], 'seat')

    ot_bileta = None
    v_bilet = None

    # Находим оба билета
    for bilet in arr[1:]:
        if bilet[id_index] == ticket_id_a:
            ot_bileta = bilet
        elif bilet[id_index] == ticket_id_b:
            v_bilet = bilet

    # Проверяем условия
    if ot_bileta and v_bilet:
        if (ot_bileta[booked_index] == 'True' and
            v_bilet[booked_index] == 'True' and
            ot_bileta[time_index] == v_bilet[time_index]):

            # Меняем местами seat и customer
            ot_bileta[seat_index], v_bilet[seat_index] = v_bilet[seat_index], ot_bileta[seat_index]
            ot_bileta[customer_index], v_bilet[customer_index] = v_bilet[customer_index], ot_bileta[customer_index]
            return True

    return False
```

```python
def find_contiguous(arr, time: str, row: str, n: int) -> list[str]:
    id_index = find_index(arr[0], 'id')
    time_index = find_index(arr[0], 'time')
    booked_index = find_index(arr[0], 'booked')
    customer_index = find_index(arr[0], 'customer')
    seat_index = find_index(arr[0], 'seat')

    seats=[bilet for bilet in arr[1:] if bilet[time_index]==time and bilet[seat_index].startswith(row)]

    seats.sort(key=lambda x: int(x[seat_index][1:]))

    for i in range(len(seats) - n + 1):
        group = seats[i:i+n]
        if all(s[booked_index] == 'False' for s in group):
            return [s[id_index] for s in group]

    return []  
```

```python
def movies_by_format(arr, format: str) -> list[str]:
    movie_index=find_index(arr[0],'movie')  
    time_index=find_index(arr[0], 'time') 
    available_index=find_index(arr[0],'booked')
    klient_index=find_index(arr[0],'customer')
    id_index=find_index(arr[0],'id') 
    price_index=find_index(arr[0],'price')
    seat_index = find_index(arr[0], 'seat')
    format_index=find_index(arr[0], 'format')
    result=[]
    for film in arr[1:]:
        if film[format_index]==format:
            if film[movie_index] not in result:
                result.append(film[movie_index])
    return result
print(movies_by_format(kino, '2D')) 
    
```

```python
def avg_price_by_format(arr) -> dict[str, float]:
    movie_index=find_index(arr[0],'movie')  
    time_index=find_index(arr[0], 'time') 
    available_index=find_index(arr[0],'booked')
    klient_index=find_index(arr[0],'customer')
    id_index=find_index(arr[0],'id') 
    price_index=find_index(arr[0],'price')
    seat_index = find_index(arr[0], 'seat')
    format_index=find_index(arr[0], 'format')
    otvet={}
    for bilet in arr[1:]:
        formatik=bilet[format_index]
        if formatik not in otvet:
            otvet[formatik]={'summa':0.0, 'count':0}
            otvet[formatik]['summa']+=float(bilet[price_index])
            otvet[formatik]['count']+=1
    for formatik in otvet:
        otvet[formatik]=otvet[formatik]['summa']/otvet[formatik]['count']
    return otvet
print(avg_price_by_format(kino))



            

    
```

```python
def min_price_format(arr) -> dict[str, float]:
    movie_index=find_index(arr[0],'movie')  
    time_index=find_index(arr[0], 'time') 
    available_index=find_index(arr[0],'booked')
    klient_index=find_index(arr[0],'customer')
    id_index=find_index(arr[0],'id') 
    price_index=find_index(arr[0],'price')
    seat_index = find_index(arr[0], 'seat')
    format_index=find_index(arr[0], 'format')
    otvet={}
    for bilet in arr[1:]:
        muvi=bilet[format_index]
        price=float(bilet[price_index])
        if muvi not in otvet:
            otvet[muvi]=price
        else:
            if price<otvet[muvi]:
                otvet[muvi]=price
    return otvet
print(min_price_format(kino))
```

```python
def apply_absolute_discount(arr, movie: str|None, format:str|None, time: str|None, amount: float) -> int:
    movie_index=find_index(arr[0],'movie')  
    time_index=find_index(arr[0], 'time') 
    available_index=find_index(arr[0],'booked')
    klient_index=find_index(arr[0],'customer')
    id_index=find_index(arr[0],'id') 
    price_index=find_index(arr[0],'price')
    seat_index = find_index(arr[0], 'seat')
    format_index=find_index(arr[0], 'format')
    count=0
    for bilet in arr[1:]:
        if (
            (bilet[movie_index]==movie or movie is None) 
            and (bilet[format_index]==format or format is None)
            and (bilet[time_index]==time or time is None)
            and (float(bilet[price_index])>= amount)
        ):
            bilet[price_index]=str(float(bilet[price_index])-amount)
           
            count+=1
    return count
```
