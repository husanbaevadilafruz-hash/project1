```python
def report(library: dict) -> dict:
    total_books = 0
    available = 0
    checked_out = 0
    sections_info = {}

    for section, books in library.items():
        section_books = 0
        section_available = 0
        section_checked_out = 0

        for book in books:
            section_books += 1
            if book["available"]:
                section_available += 1
            else:
                section_checked_out += 1

        sections_info[section] = {
            "total": section_books,
            "available": section_available,
            "checked_out": section_checked_out
        }

        total_books += section_books
        available += section_available
        checked_out += section_checked_out

    return {
        "total_books": total_books,
        "available": available,
        "checked_out": checked_out,
        "sections": sections_info
    }

print(report(library))
```
