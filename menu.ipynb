```python
def count_items(menu: dict) -> int:
    count=0
    for foods in menu.values():            #kolichestvo blud
        for food in foods:
                 count+=1
    return count
print(count_items(menu))
```

```python
def count_orderable(menu: dict) -> int:
    count_available=0
    for foods in menu.values():
        for food in foods:
            if food["available"]==True:
                count_available+=1
    return count_available
print(count_orderable(menu))
```

```python
def find_category(menu: dict, item_id: str) -> str | None:
    for category, foods in menu.items():
        for food in foods:
            if food["id"].casefold()==item_id.casefold():   #naiti categoriu bluda
                return category 
    return None
print(find_category(menu, "m03"))
```

```python
def booook(menu: dict, item_id: str) -> str | None:
    for category, foods in menu.items():
        for food in foods:
            if food["id"].casefold()==item_id.casefold():  #vernut slovar bluda
                return food
    return None
print(booook(menu, 'm03'))
```

```python
def restock(menu: dict, item_id: str, amount: int) -> bool:
    for foods in menu.values():
        for food in foods:
            if food["id"].casefold()==item_id.casefold():
                food["stock"]+=amount
                if food["stock"]>0:
                    food["available"]=True                  #popolnit zapas
                    return True
    return False
print(restock(menu, "s01", 5))  
print(menu)
```

```python
def set_available(menu: dict, item_id: str, is_available: bool) -> bool:
    for foods in menu.values():
        for food in foods:
            if food["id"].casefold()==item_id.casefold():
                food["available"]=is_available              #pomenat available
                return True
    return False
print(set_available(menu, 's01', False))
print(menu)
```

```python
def remove_item(menu: dict, item_id: str) -> bool:
    for foods in menu.values():
        for food in foods:
            if food["id"].casefold()==item_id.casefold():
                foods.remove(food)
                return True                                 #udalit bludo po id
    return False
print(remove_item(menu, 's01' ))
print(menu)
```

```python
def update_price(menu: dict, item_id: str, new_price: float) -> bool:
    if new_price<0:
        return False
    for foods in menu.values():
        for food in foods:                                #pomenat cenu
            if food["id"].casefold()==item_id.casefold():
                food["price"]=new_price
                return True
    return False
print(update_price(menu,'s02',56))
print(menu)
```

```python
def add_item(menu: dict, category: str, item: dict) -> bool:
   if category  not in menu:
      menu[category]=[] 

   for food in menu[category]:
         if food["id"].casefold()==item["id"].casefold():  #dobavit bludo
            return False                                   
   menu[category].append(item)
   return True 
print(add_item(menu, "startes", {'id': 'S90', 'name': 'Chicken', 'price': 89, 'stock': 45, 'available': True, 'veg': False}))
print(menu)
      
```
