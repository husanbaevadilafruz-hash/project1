```python
from random import randint
count = 0
coin = ["orel", "reshka"]
lastindex = len(coin) - 1
for _ in range(100):
    a = randint(0, lastindex)
    if a == "orel":
        count += 1
    count2 = count / 100
print(count2 / 3)
```

```python
from random import randint
co = 0
for _ in range(300):
    a = randint(1, 2)
    b = randint(1, 2)
    c = randint(1, 2)
    if a == b == c:
        co += 1
print("Probability:", co / 300)
```

```python
from random import randint
coin = ['orel', 'reshka']
lastindex = len(coin) - 1
co = 0
for _ in range(300):
    a = randint(0, lastindex)
    first = coin.pop(a)
    coin = ['orel', 'reshka']
    if first == 'orel':
        co += 1
co2 = co / 300
print("Probability:", co2 / 3)
```
