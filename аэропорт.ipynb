```python
def count_seats(seating:dict)->int:
    count=0
    for classes in seating.values():            #obshee kolichestvo mest
        for seat in  classes:
            count+=1
    return count
print(count_seats(seating))
```

```python
def count_occupied_seats(seating:dict)->int:
    count=0
    for classes in seating.values():            
        for seat in  classes:                  #kolichestvo zabronirovannyh mest
            if seat["occupied"]==False:
                count+=1
    return count
print(count_occupied_seats(seating))
```

```python
def find_seat_class(seating: dict, seat_id: str) -> str | None:
    for klass, seats in seating.items():                           # перебираем все классы
        for seat in seats:                                      # перебираем места в каждом классе
            if seat["id"].casefold() == seat_id.casefold():
                return klass                                  # возвращаем класс, если нашли
    return None                                              # если не нашли

print(find_seat_class(seating, "6C"))
```

```python
def assign_seat(seating: dict, seat_id: str, passenger: str) -> bool:
    for seats in seating.values():
        for seat in seats:
           if seat["id"].casefold()==seat_id.casefold():
               if not seat["occupied"]:    
                   seat["occupied"]=True                       #naznachit mesto                       
                   seat["passenger"]=passenger
                   return True
               else:
                   return False
               
    return False
    
print(assign_seat(seating,  "1c","dilia"))
```

```python
def release_seat(seating: dict, seat_id: str) -> bool:
    for seats in seating.values():
        for seat in seats:
            if seat["id"].casefold() == seat_id.casefold():
                if seat["occupied"]:                          #osvobodit mesto
                    seat["occupied"] = False
                    seat["passenger"] = None
                    return True
    return False

# Проверка
print(release_seat(seating, "12B"))
print(seating)
```

```python
def list_available(seating: dict) -> list[str]:
  spisokk=[]
  for seats in seating.values():
    for seat in seats:
        if not seat["occupied"]:                                       #spisok svobodnyh mest
            spisokk.append(seat["id"])
  return spisokk
print(sorted(list_available(seating),key=str.casefold))
    
```

```python
def section_size(seating: dict, klass: str) -> int:
    for key, seats in seating.items():
        if key.casefold()==klass.casefold():             #kolichestvo mest v spiske
            return len(seats)
        
    return 0
print(section_size(seating,"business"))
```

```python
def section_size(seating: dict, klass: str) -> int:
    seats = seating.get(klass, [])                   #kolichestvo mest v spiske
    return len(seats) 
print(section_size(seating, "Business"))
```

```python
def longest_passenger(seating: dict) -> str | None:
    longest_name=None
    for seats in seating.values():
        for seat in seats:
            if seat["passenger"]:
                if (
                longest_name is None
                or len(seat["passenger"])>len(longest_name)
                or (len(seat["passenger"])==len(longest_name) 
                and seat["passenger"].casefold()<longest_name.casefold())
                ):
                    longest_name=seat["passenger"]
    return longest_name
print(longest_passenger(seating))

```

```python
def longest_passenger(seating: dict) -> str | None:
    longest_name = None
    for seats in seating.values():
        for seat in seats:
            if seat["passenger"]:  # проверяем, что место занято
                if (
                    longest_name is None
                    or len(seat["passenger"]) > len(longest_name)
                    or (len(seat["passenger"]) == len(longest_name)
                        and seat["passenger"].casefold() < longest_name.casefold())
                ):
                    longest_name = seat["passenger"]
    return longest_name

print(longest_passenger(seating))
```
